```python
def place_move(board, player, x, y):
    # first check: are the coordinates even inside the board?
    # if someone clicks outside the board, thatâ€™s not allowed
    if x < 0 or y < 0 or x >= len(board) or y >= len(board):
        return False

    # now check if the spot is empty
    # we only allow placing something if the value is 0
    if board[x][y] != 0:
        return False   # spot already taken

    # decide what symbol to place
    # player 1 gets X, player 2 gets O
    if player == 1:
        board[x][y] = "X"
    elif player == 2:
        board[x][y] = "O"
    else:
        # just in case someone passes a weird player number
        return False

    # if we reached this point, the move worked
    return True
    
```

```python
def draw_board(board):
    n = len(board)          # number of rows
    m = len(board[0])       # number of columns

    # print column labels (A, B, C, ...)
    print("  ", end="")
    for col in range(m):
        print(" " + chr(65 + col) + " ", end="")
    print()

    # now print each row
    for i in range(n):
        # print row number (1, 2, 3, ...)
        print(i + 1, end=" ")

        for j in range(m):
            if board[i][j] == 0:
                print(" ", end="")
            else:
                print(board[i][j], end="")

            # print vertical lines between columns
            if j < m - 1:
                print(" | ", end="")
        print()

        # print horizontal lines between rows
        if i < n - 1:
            print("  " + "---+" * (m - 1) + "---")
```

```python
def place_move_label(board, player, location):
    # location is something like "A2"

    # get column letter (A, B, C...)
    col_letter = location[0]

    # get row number (1, 2, 3...)
    row_number = location[1]

    # convert column letter to index
    # ord("A") is 65, so subtract 65 to get 0
    col = ord(col_letter) - 65

    # convert row number to index
    # subtract 1 because humans start at 1, python starts at 0
    row = int(row_number) - 1

    # now call exercise 5 to actually place the move
    return place_move(board, player, row, col)
```

```python
def player_turn(board, player):
    valid_move = False   # we start assuming the move is NOT valid

    while not valid_move:
        # ask the player where they want to play
        location = input("Enter your move (like A2): ")

        # try to place the move using exercise 7
        valid_move = place_move_label(board, player, location)

        # if the move didn't work, tell them and try again
        if not valid_move:
            print("That move is not valid. Try again.")
```

```python
def play_tic_tac_toe():
    # create a 3x3 empty board
    board = [
        [0, 0, 0],
        [0, 0, 0],
        [0, 0, 0]
    ]

    game_status = -1   # -1 means the game is still going
    current_player = 1

    # keep playing while the game is not over
    while game_status == -1:
        draw_board(board)

        print("Player", current_player, "turn")
        player_turn(board, current_player)

        # check if someone won or if it's a draw
        game_status = check_game(board)

        # switch players
        if current_player == 1:
            current_player = 2
        else:
            current_player = 1

    # game is over, draw final board
    draw_board(board)

    # print result
    if game_status == 1:
        print("Player 1 wins!")
    elif game_status == 2:
        print("Player 2 wins!")
    else:
        print("It's a draw!")
        
```
