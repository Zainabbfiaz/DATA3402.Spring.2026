1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
  
    def __init__(self,max_value):
        self.value = 0 
        self.max_value = max_value

    
    def increment(self):
        if self.value < self.max_value:
            self.value += 1
            print(self.value)
        else: 
            print('Error:Counter reached maximum value')

    def reset(self):
        self.value = 0 
        print('counter reset to zero')
            
    
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class shape:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    def contains(self, px, py):
        raise NotImplementedError("Subclass must implement contains")

class circle(shape):
    def __init__(self, x, y, radius):
        super().__init__(x, y)
        self._radius = radius

    def contains(self, px, py):
        dx = px - self._x
        dy = py - self._y
        return dx * dx + dy * dy <= self._radius * self._radius

class rectangle(shape):
    def __init__(self, x, y, length, width):
        super().__init__(x, y)
        self._length = length
        self._width = width

    def contains(self, px, py):
        if px >= self._x and px <= self._x + self._length:
            if py >= self._y and py <= self._y + self._width:
                return True
        return False

class triangle(shape):
    def __init__(self, x, y, base, height):
        super().__init__(x, y)
        self._base = base
        self._height = height

    def contains(self, px, py):
        
        if px < self._x or px > self._x + self._base:
            return False
        if py < self._y or py > self._y + self._height:
            return False

        # Check if below diagonal
        # equation of diagonal from (x+base,y) to (x,y+height)
        slope = -self._height / self._base
        max_y = slope * (px - self._x) + self._height

        if py <= self._y + max_y:
            return True

        return False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
def overlaps(self, other):
    
    my_points = self.get_points()
    for p in my_points:
        if other.contains(p[0], p[1]):
            return True

    
    other_points = other.get_points()
    for p in other_points:
        if self.contains(p[0], p[1]):
            return True

    return False
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
